# Data Preprocessing

## Handling Missing Values

```python
print(df.isnull().sum())
```

```python
df.duplicated().sum()
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler

obj_cols = df.select_dtypes(object).columns
for col in df.columns:
  if col not in obj_cols:
    scaler = StandardScaler()
    df[col] = scaler.fit_transform(df[[col]])
```

```python
df.dtypes
```

## Encode Categorical Variables

```python
from sklearn.preprocessing import LabelEncoder
obj_cols = df.select_dtypes(object).columns
for col in obj_cols:
  encoder = LabelEncoder()
  df[col] = encoder.fit_transform(df[col])

df.dtypes
```

# Split the Data

```python
x = df.drop(columns = ['Target'])
y = df['Target']
```

```python
from sklearn.model_selection import train_test_split

xtrain, xtest, ytrain, ytest = train_test_split(x, y, train_size = 0.8)
```

# Model Selection

```python
from sklearn.metrics import accuracy_score

def eval_model(model, xtrain, ytrain, xtest, ytest):
  model.fit(xtrain, ytrain)
  trainpred = model.predict(xtrain)
  testpred = model.predict(xtest)
  return accuracy_score(ytrain, trainpred), accuracy_score(ytest, testpred)
```

```python
result = pd.DataFrame(columns = ['Model Number', 'Name', 'Training Accuracy', 'Testing Accuracy'])
result.head()
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
row = []
row.extend([1, 'Logistic Regression'])
row.extend(eval_model(model, xtrain, ytrain, xtest, ytest))
result.loc[len(result.index)] = row
result.head()
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
row = []
row.extend([2, 'Decision Tree'])
row.extend(eval_model(model, xtrain, ytrain, xtest, ytest))
result.loc[len(result.index)] = row
result.head()
```

```python
model = DecisionTreeClassifier(max_depth = 10)
eval_model(model, xtrain, ytrain, xtest, ytest)
```

```python
model = DecisionTreeClassifier(min_samples_leaf = 10)
eval_model(model, xtrain, ytrain, xtest, ytest)
```

```python
model = DecisionTreeClassifier(max_leaf_nodes = 10, class_weight = 'balanced')
eval_model(model, xtrain, ytrain, xtest, ytest)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier()
row = []
row.extend([3, 'KNN'])
row.extend(eval_model(model, xtrain, ytrain, xtest, ytest))
result.loc[len(result.index)] = row
result.head()
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(max_depth = 10)
row = []
row.extend([4, 'Decision Tree Classifier'])
row.extend(eval_model(model, xtrain, ytrain, xtest, ytest))
result.loc[len(result.index)] = row
result.head()
```

```python
import numpy as np

models = result['Name']
training_accuracies = np.round(result['Training Accuracy'],4)
testing_accuracies = np.round(result['Testing Accuracy'],4)

x = np.arange(len(models))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots(figsize=(10, 6))
rects1 = ax.bar(x - width/2, training_accuracies, width, label='Training Accuracy')
rects2 = ax.bar(x + width/2, testing_accuracies, width, label='Testing Accuracy')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_xlabel('Model')
ax.set_ylabel('Accuracy')
ax.set_title('Training and Testing Accuracy of Various ML Models')
ax.set_xticks(x)
ax.set_xticklabels(models)
ax.legend()

# Function to add labels on top of the bars
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(4, 4),  # 4 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)

fig.tight_layout()

plt.show()
```

# Model Training

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(max_depth = 10)
model.fit(xtrain, ytrain)
```

# Model Evaluation

```python
trainpred = model.predict(xtrain)
testpred = model.predict(xtest)
```

```python
from sklearn.metrics import classification_report

print(classification_report(ytrain, trainpred))
```

```python
print(classification_report(ytest, testpred))
```
